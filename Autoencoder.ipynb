# Goal
The goal of this notebook is to get familiar with autoencoders and works as an introduction to implementing a variational autoencoder. Roughly the most simple idea about an autoencoder is the presence of $2$ parts, an encoder with the objective of reducing the dimensions of the inputs (compression) and a decoder with the objective of reconstructing the inputs from the compression.

# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
```

# Prepare the data

```python
# Detect if we have a GPU available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Preprocessing
transform = torchvision.transforms.Compose([torchvision.transforms.ToTensor()])

# Import the training dataset
train_dataset = torchvision.datasets.MNIST(
    root="~/torch_datasets", train=True, transform=transform, download=False)

# Import the test dataset
test_dataset = torchvision.datasets.MNIST(
    root="~/torch_datasets", train=False, transform=transform, download=False)

# Create a dataloader for the train data
train_loader = torch.utils.data.DataLoader(
    train_dataset, batch_size=32, shuffle=True, num_workers=0, pin_memory=True)

# Create a dataloader for the test data
test_loader = torch.utils.data.DataLoader(
    test_dataset, batch_size=128, shuffle=False, num_workers=0)

dataloaders = {'train':train_loader, 'test':test_loader}
```

# Create the model (Autoencoder)

```python
class AutoEncoder(nn.Module):
    def __init__(self, **kwargs):
        super(AutoEncoder, self).__init__()
        self.encoder_hidden_layer = nn.Linear(in_features=kwargs['input_shape'], out_features=128)
        self.encoder_output_layer = nn.Linear(in_features=128, out_features=128)
        self.decoder_hidden_layer = nn.Linear(in_features=128, out_features=128)
        self.decoder_output_layer = nn.Linear(in_features=128, out_features=kwargs['input_shape'])
        
    def forward(self, x):
        x = F.relu(self.encoder_hidden_layer(x))
        x = F.relu(self.encoder_output_layer(x))
        x = F.relu(self.decoder_hidden_layer(x))
        x = F.relu(self.decoder_output_layer(x))
        return x
```

# Create a training function

```python
def train(model, dataloaders, device, optimizer, criterion, num_epochs=10):
    for epoch in range(num_epochs):
        print('----------------')
        print('Epoch: ', epoch)
        print('----------------')
        
        running_loss = 0
        
        for phase in ['train', 'test']:
            for inputs,_ in dataloaders[phase]:
                inputs = inputs.view(-1,784).to(device)
            
                optimizer.zero_grad()
            
                with(torch.set_grad_enabled(phase == 'train')):
                    outputs = model(inputs)
                    loss = criterion(outputs, inputs)
                
                    if phase=='train':
                        loss.backward()
                        optimizer.step()
                        
                running_loss += loss.item()
                
            print('{} Loss: {:.4f}'.format(phase, running_loss))
```

# Train the model

```python
# Instanciate the model        
model = AutoEncoder(input_shape=784).to(device)

# Create an optimizer
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Create a loss fxn
criterion = nn.MSELoss()

train(model, dataloaders, device, optimizer, criterion)
```

# Vizualize the results

```python
# Try the model on some test images
samples_idx = np.random.randint(1000, size=10)
samples_subset = torch.utils.data.Subset(test_dataset, samples_idx)
samples_loader = torch.utils.data.DataLoader(samples_subset, len(samples_idx))

with torch.no_grad():
    for inputs,_ in samples_loader:
        test_examples = inputs.view(-1,784)
        reconstructed = model(test_examples)

# we want a figure with 2 rows and 10 columns, the upper row contains the test examples
# and the lower row contains the reconstructed examples
h = 28
w = 28
cols = 10
rows = 2
fig = plt.figure(figsize=(30, 10))
for i in range(1, cols+1):
    img = test_examples[i-1].reshape(28,28)
    fig.add_subplot(rows, cols, i)
    plt.imshow(img)
for i in range(1, cols+1):
    img = reconstructed[i-1].reshape(28,28)
    fig.add_subplot(rows, cols, i+cols)
    plt.imshow(img)
plt.show()
```

# Different compression levels

```python
class AutoEncoder_v2(nn.Module):
    def __init__(self, **kwargs):
        super(AutoEncoder_v2, self).__init__()
        self.encoder_hidden_layer = nn.Linear(in_features=kwargs['input_shape'], out_features=128)
        self.encoder_hidden_layer2 = nn.Linear(in_features=128, out_features=64)
        self.encoder_output_layer = nn.Linear(in_features=64, out_features=32)
        self.decoder_hidden_layer = nn.Linear(in_features=32, out_features=64)
        self.decoder_hidden_layer2 = nn.Linear(in_features=64, out_features=128)
        self.decoder_output_layer = nn.Linear(in_features=128, out_features=kwargs['input_shape'])
        
    def forward(self, x):
        x = F.relu(self.encoder_hidden_layer(x))
        x = F.relu(self.encoder_hidden_layer2(x))
        x = F.relu(self.encoder_output_layer(x))
        x = F.relu(self.decoder_hidden_layer(x))
        x = F.relu(self.decoder_hidden_layer2(x))
        x = F.relu(self.decoder_output_layer(x))
        return x
```

```python
# Instanciate the model        
model2 = AutoEncoder_v2(input_shape=784).to(device)

# Create an optimizer
optimizer2 = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

# Train
train(model, dataloaders, device, optimizer2, criterion)
```

# Convolutional Autoencoder
After doing some research, I found a neural network architecture called Convolutional Autoencoder which contains convolutional layers, let's try it.

# Create the model

```python
class ConvAutoEncoder(nn.Module):
    def __init__(self):
        super(ConvAutoEncoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 16, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, 7)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 32, 7),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 1, 3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

# Training the model
we will write a new training method just for the sake of practicing and optimizing our method

```python
# This time we will create the optimizer and the criterion in the training function
def train_v2(model, train_dataset, val_dataset, num_epochs=10, batch_size=64, learning_rate=0.001):
    # Create the optimizer and a loss fxn
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    criterion = nn.MSELoss()
    
    # Detect if we have a GPU and send the model to the GPU
    device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    
    # Create a dataloader
    train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_dataloader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    dataloaders = {'train': train_dataloader, 'val': val_dataloader}
    
    for epoch in range(num_epochs):
        print('----------')
        print('epoch: ',epoch)
        print('----------')
        
        logs = []
        for phase in ['train', 'val']:
            running_loss = 0
            for inputs,_ in dataloaders[phase]:
                inputs = inputs.to(device)
                # Enable gradient requirement only in training mode
                with torch.set_grad_enabled(phase=='train'):
                    reconstruct = model(inputs)
                    loss = criterion(inputs, reconstruct)
                    running_loss += loss.item()
                    # We backprop only in training
                    if phase=='train':
                        loss.backward()
                        optimizer.step()
                    # We reset the gradient in both train and val just to be cautious
                    optimizer.zero_grad()
                # We append the current results to see later how the model improved
                logs.append((epoch, inputs, reconstruct))
                
            print('{} Loss: {:.4f}'.format(phase, running_loss))
            
    return logs
```

# Train the model

```python
# Initialize the model
model_conv_ae = ConvAutoEncoder()

# Train the model
logs = train_v2(model_conv_ae, train_dataset, test_dataset)
```

# Visualize the improvement of the model throughout training

```python
for k in range(0, 10, 5):
    plt.figure(figsize=(9,2))
    imgs = logs[k][1].detach().numpy()
    reconst = logs[k][2].detach().numpy()
    for i, item in enumerate(imgs):
        if i>=9: break
        plt.subplot(2, 9, i+1)
        plt.imshow(item[0])
        
    for i, item in enumerate(reconst):
        if i>=9: break
        plt.subplot(2, 9, 9+i+1)
        plt.imshow(item[0])
```
